```python
import math
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
p_num = 101
vs = np.logspace(1, 7, p_num)
```

```python
Ps = np.empty(p_num)
```

```python
for i in range(p_num):
    z,p,k = signal.tf2zpk([0, 1], [vs[i], 1])
Ps[i] = p[0]
```

```python
fg,ax = plt.subplots()
ax.plot(np.real(Ps), np.imag(Ps), '.-')
for i in range(0,p_num,10):
    print('For a=={0:.2f} P==({1:.2f}, {2:.2f})'.format(vs[i],Ps[i].real, Ps[i].imag))
plt.show()
```

```python
freq = 0.1
T = np.linspace(0, 200.0/freq, 10001)
a = 3
model = signal.tf2zpk([0, 1], [a, 1])
S = [1 for t in T[:2001]]
Tout, yout, xout = signal.lsim(model, S, T[:2001])
plt.plot(Tout[:2001], yout[:2001])
plt.title("Реакция на единичный сигнал при a=3")
plt.figure()
```

```python
freq = 0.1
T = np.linspace(0, 200.0/freq, 10001)
a = -3
model = signal.tf2zpk([0, 1], [a, 1])
S = [1 for t in T[:2001]]
Tout, yout, xout = signal.lsim(model, S, T[:2001])
plt.plot(Tout[:2001], yout[:2001])
plt.title("Реакция на единичный сигнал при a=-3")
plt.figure()
```

```python
a = 3
model = signal.tf2zpk([0,1], [a, 1])
Omegas, mag, phase = signal.bode(model)
fs = [0.5*w/math.pi for w in Omegas]
plt.figure()
plt.semilogx(fs, mag) 
plt.title("АЧХ, устойчивая(a=3)")
```
